```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
from IPython.display import Image
import matplotlib.pyplot as plt
import numpy as np
import json
from scipy.stats import scoreatpercentile
from overview_maps_reakt import main as delaymap
from magnitude_comparison import MagComp
from optimal_blindzone import plot_bz
import sys
sys.path.append('../reports')
from point_in_polygon import EventSoCal, EventCA

def plot_pick_delays(fin, fout=None, networks=['*'],
                     lineplot='False'):
    fh = open(fin)
    pkdel = json.load(fh)
    fh.close()
    delays = {}
    for _k in pkdel:
        net,sta = _k.split('.')
        if net not in networks and '*' not in networks:
            continue
        if net not in delays:
            delays[net] = pkdel[_k]
        else:
            delays[net] += pkdel[_k]
    fig = plt.figure()
    ax = fig.add_subplot(111)
    if lineplot:
        all_delays = []
        for _n in delays.keys():
            all_delays += delays[_n]
        n, bins, patches = ax.hist(all_delays,bins=np.arange(0, 20,1.0),
                               histtype='step', color='k',
                               rwidth=1.0, normed=True)
    else:
        n, bins, patches = ax.hist([delays[_n] for _n in delays.keys()],
                                   bins=np.arange(-30, 30,1.0),
                                   histtype='barstacked',
                                   label=[_n for _n in delays.keys()],
                                   rwidth=1.0)

    ax.set_xlabel('Pick delays [s]')
    ax.legend()
    if fout is not None:
        plt.savefig(fout, dpi=300)

def plot_origin_creation_time(fin,fout=None):
    a = np.load(fin)
    delays_ct = a['delays']
    fig = plt.figure()
    ax1 = fig.add_subplot(111)
    n, bins, patches = ax1.hist(delays_ct, bins=np.arange(-30, 30, 1.0),
                                color='blue', rwidth=1.0)
    med = np.median(delays_ct)
    percentile16 = scoreatpercentile(delays_ct, 16)
    percentile84 = scoreatpercentile(delays_ct, 84)
    ax1.text(0.6, 0.7, 'Median: %.1f s' % (med), horizontalalignment='left',
            transform=ax1.transAxes, color='blue')
    ax1.text(0.6, 0.65, '16th percentile: %.1f s' % (percentile16), horizontalalignment='left',
            transform=ax1.transAxes, color='blue')
    ax1.text(0.6, 0.6, '84th percentile: %.1f s' % (percentile84), horizontalalignment='left',
            transform=ax1.transAxes, color='blue')
    ax1.set_xlabel('Event declaration time [s]')
    if fout is not None:
        plt.savefig(fout, dpi=300)
```

## The Virtual Seismologist in SeisComP3 as an implementation model for Earthquake Early Warning algorithms

### Introduction
Earthquake Early Warning (EEW) describes the concept of using the first observations of seismic energy at sensors closest to the rupture to predict expected ground motion levels at further distant sites. This can either be done by first inferring rupture parameters such as hypocenter and rupture extent and then predicting ground motion levels based on a ground motion prediction equation and distance to the rupture [references VS, Elarms, PRESTo, Onsite, FinDer] or by directly computing expected peak ground motion either empirically [reference PRESTo-onsite] or using physical models of wavefield propagation [reference Hoshiba]. The principles of EEW were already described by Cooper [] and Heaton [1985] but first implementations were not realized until ??. Now, several regions operate (e.g. Japan, Mexico, Romania) or test EEW systems (e.g. China, Italy, Taiwan, Turkey, Greece, several states on the US and Canadian west coast). 

The debate whether earthquake rupture is, at least in parts, deterministic [Olson and Allen, Colombelli et al.], or whether rupture size is fully determined by the distribution of stress and material properties on the fault [Wesnousky] is still ongoing. In practice, EEW systems have been developing into multi-algorithm systems, using different strategies to estimate rupture parameters and peak ground motion depending on the amount and type of data available (e.g. GPS and seismic waveforms) during or immediately following a rupture and are therefore somehow forestalling the outcome of the scientific debate. Japan is already operating such a multi-algorithm system, while California and Italy are still testing theirs.  

A natural consequence of such an EEW approach is to regard earthquake monitoring as a progression from fast EEW estimates of earthquake rupture parameters associated with large uncertainties to a detailed analysis with small uncertainties hours later. Such systems are, however, more complex and therefore more challenging to operate than an EEW system based on one algorithm only. Information from different algorithms has to be combined consistently which requires additional logic but also implementation standards that make output form different algorithms compatible. On the example of the Virtual Seismologist EEW algorithm we describe how such a system can be realized within the SeisComP3 (SC3) earthquake monitoring platform (VS(SC3)), which becomes increasingly popular with seismic network operators world wide. While our implementation here is particular to SC3, similar solutions could be implemented in other real-time monitoring platforms.

Integrating EEW into routine earthquake monitoring systems further facilitates operating EEW algorithms in seismic networks which do not have dedicated EEW research groups. This is of increasing relevance as seismic network densities and capabilities in many earthquake prone regions are continuously improving in a way that may make EEW a reality within the next decade. Providing operators of these networks with EEW solutions that are configured and run the same way as other earthquake monitoring modules will significantly lower the additional effort required to provide EEW alerts. We show first results of running VS(SC3) in seven real-time seismic networks around the globe and will outline further necessary algorithmic and infrastructural improvements to minimise alert delays in these networks.



### The Virtual Seismologist in SeisComP3 (VS(SC3))

SC3 is an open-source earthquake monitoring system with professional software support (gempa GmbH). It follows a modular approach where every module is a standalone program that communicates with other modules through a TCP/IP messaging system (spread [reference]) and connects to a common database containing, amongst other information, the station meta data. The messaging system facilitates the setup of systems distributed over several computing nodes that can be situated in different locations. The modular approach allows running and testing single components independently from the whole monitoring infrastructure but also enables running different algorithms receiving the same input, easily replacing modules, and running multiple instances of the same algorithm with different configuration settings. As datamodel, SC3 uses a derivative of QuakeML [reference], which has become the standard for earthquake parameter exchange in recent years. This datamodel is used to store output from different modules in the central database but it also enables an easy exchange with other systems capable of parsing QuakeML. We will describe one such example in the following section. QuakeML is a comprehensive, flexible, and extensible datamodel that allows to store any data related to an earthquake detection. For real-time operations waveforms are provided by one or more (??is that true??) SeedLINK servers and for subsequent analysis the same waveforms can be requested through FDSN web services [reference], ArcLink [reference], or local waveform repositories. SC3 also comes with a sophisticated Graphical User Interface that was designed for routine earthquake monitoring. By providing output that is compatible with the SC3 datamodel, this GUI can also be used to analyze the performance of VS(SC3). 

SC3 also comes with an application programming interface (API) in C++ and Python which facilitates the use of the SC3 messaging system and provides access to common math operations and to seismic waveforms. No changes to the source code are required in order to receive waveforms from archives or a SeedLINK server which ensures a seamless transition from off-line testing of prototype implementations to operating them in real-time.

VS(SC3) implements the VS likelihood function [Cua, 2006; Cua and Heaton, 2007; Cua et al., 2009] (see appendix) which computes the most likely magnitude and epicentral location given maximum envelope observations on the root-mean-square horizontal component and the vertical component of the acceleration, velocity and displacement stream following a P-wave detection. In the current implementation, VS(SC3) receives location estimates from SC3's standard location module (scautoloc) and then reduces the original 3-dimensional likelihood function (magnitude, latitude, longitude) to a 1-dimensional likelihood function for magnitude only. It thereby replaces VS' amplitude based strong motion centroid [reference Cua, Kanamori] with a location based on P-wave traveltimes. Detections of P-wave arrivals are provided by SC3's scautopick module which implements an STA/LTA picker [reference Allen] (mention the Baer-Kradolfer picker?). Ideally one would combine the location estimates based on amplitudes and P-wave traveltimes, however, this requires the probability density function for both location estimates which is currently only available for VS' amplitude based location estimate. We will discuss this further in Section ??. The VS likelihood function uses different parameters for P- and S-waves and tries to distinguish between both through a weighted ratio of the horizontal and vertical ground motion (see Appendix for more detail).

A separate module (scenvelope) continuously computes a stream of envelope values over 1 s long intervals after correcting the raw waveforms for gain and high-pass filtering at 0.33 Hz with a 4th order butterworth filter, to reduce the influence of microseisms on the magnitude estimates of small events. Waveforms are also checked for saturation and are marked accordingly by comparing the raw counts with the possible counts on a 24-bit digitizer. A second module (scvsmag) receives these envelope values through the messaging system and once it receives a location estimate it uses them to compute the maximum, variance, and the expected value of the VS' likelihood function for magnitude. Magnitude estimates are updated in 1 s intervals for a configurable amount of time including additional envelope values and location updates as they become available. If waveforms from a broad-band sensor have saturated, those of a co-located strong-motion sensor are used instead, if available.

A third module (scvsmaglog) receives location and VS magnitude estimates and forwards them to as a continuous stream of alerts to EEW specific user interfaces such as the UserDisplay [reference] or the recently developed Earthquake Early Warning Display (EEWD) for Europe. Internally, scvsmaglog converts the SC3 datatype messages to either QuakeML (required by the EEWD) or ShakeAlert's custom XML format (required by the UserDisplay) using Extensible Stylesheet Language Transformations (XSLT) and then sents them to an ActiveMQ broker [reference] which forwards the messages to subscribed end-users. The use of XSLTs facilitates the addition of other alert message formats. Through SC3's modular approach several instances of scvsmaglog can be run each receiving the same input but sending alert messages in different formats, thereby providing input for both the UserDisplay and the EEWD. scvsmaglog can further send emails and store alert histories on disk. Figure 1 shows a schematic discription of the VS(SC3) modules and their information exchange within the SC3 environment.



### Performance of VS(SC3) worldwide

Since its implementation in 2013, VS(SC3) has been installed at seven seismic networks worldwide (Switzerland, Greece, New Zealand, Romania, Turkey, Iceland, southern California). This was largely facilitated through SC3's dissemination strategy, where gempa provides pre-compiled versions of SC3 for many different Linux platforms. Network operators already running SC3 for routine earthquake monitoring can therefore operate VS(SC3) by simply updating their SC3 installation to a version containing the VS(SC3) modules. Configuration of the VS(SC3) modules is supported through extensive documentation, which is also embedded in SC3's standard documentation [reference URL].

Seismicity within these seven networks differs strongly ranging from interplate earthquakes at subduction zones (Greece, New Zealand), continental plate boundaries (California, New Zealand, Turkey, Greece, Switzerland), and mid-ocean ridges (Iceland) to intermediate-depth sub-crustal seismicity (Romania) and earthquakes related to volcanic activity (Iceland, New Zealand) [list of references on tectonic settings/seismicity]. Including VS into SeisComP3 has given us the unique opportunity to test an EEW algorithm in real-time within these different tectonic settings with a wide range of source mechanisms, magnitudes and hypocentral depths. All regions are exposed to moderate or high seismic hazard [reference share, USGS hazard maps, NZ seismic hazard program] and could therefore profit from an operational EEW system.

Figure 2 shows the comparison of the first magnitude estimate provided by VS(SC3) (MVS) with the final local magnitude (Ml) for over 3000 events detected and located by the test networks during different periods between ?? and ?? with magnitudes between ?? and ??. For shallow seismicity MVS scales well with Ml showing the median around ?? and the 16th and 84th percentile at ?? and ??. MVS and Ml agree best for magnitudes equal or larger than Ml3.0 and location errors less than 100 km. Since the VS likelihood function does not include hypocentral depth, magnitudes of deep events are systematically underestimated. It is, however, important to note here that hypocentral depth for deep events is less well constrained than for shallow events. Consequently, the uncertainty of Ml may also be larger. Further, differences exist between definitions of Ml at different networks. We corrected both MVS and Ml where systematic differences to SC3's standard Ml implementation were known. Another class of events for which MVS tends to be inaccurate are earthquakes with unusually low frequency content such as seen in the sequence of events caused by the 2014 erruption of Bardarbunga volcano in Iceland (see Figure 2). Here the high-pass filter at 0.33 Hz also removes a considerable part of the seismic energy leading to systematically underestimated magnitudes.

Given the fact that the GMPEs for P-wave amplitude and peak ground motion used in the VS likelihood function are mostly based on southern California seismicity with the addition of the NGA-West dataset [NOTE: check in Georgia's thesis], it may seem surprising that MVS also performs well in other regions. In Figure S1 [NOTE: electronic supplemant showing VS GMPE vs. other global GMPEs] we show that VS' GMPE for peak ground motion, is comparable to and falls within the standard deviation of other global GMPEs for peak ground motion. It can therefore be expected that also the GMPE for P-wave amplitude is applicable to other areas. 

Later MVS estimates tend to be more accurate as they include more waveform data from a larger number of sensors. As a consequence, the first MVS estimate in seismic networks which send large data packets become available later but often with higher accuracy. #### Alert times

Besides fast EEW algorithms the most important factor to deliver timely earthquake alerts are dense seismic networks and fast and reliable data communication paths. Figure 4 shows that in most networks (with the exception of California and Switzerland) even in areas with high station density (P-wave traveltime to the first 6 stations <= 5 s) the first alerts are usually not issued until another 5 s after the P-wave should have reached the 6th station. This results in blind zones that are greater than 50 to 60 km. Using the relation between magnitude, source-to-site distance, and intensity by Allen et al. (2012) we estimate that only for magnitudes >= M7.4 would the area of strong shaking and moderate damage (IMM VII) be larger than the blind-zone and therefore allow a useful, albeit short, warning. This is, of course, only a first-order approximation as finite rupture extent, rupture complexity, and site and path-effects are likely to produce laterally varying ground motion intensities. It shows, however, that these warning times are too slow to send usefule alerts for moderate size earthquakes (magnitudes <= M6.5) for which point-source algorithms such as VS are best suited.

Due to source and wave-propagation effects or high noise levels on some sensors, the first VS estimate may not be based on P-wave detections at the closest six seismic stations. Unless located on a nodal plane, permanent seismic stations have typically low enough noise levels that they are unlikely to miss the P-wave arrival of earthquakes that may be felt and for which earthquakes alerts are relevant [M>=2.5?;reference Cauzzi and Clinton; Clinton]. Because every seismic parameter, such as phase detections, location or magnitude estimates, are stored in the SeisComP3 database with the timestamp of its creation, the database serves as a precise, chronological log of every event detection. This does not only allow to measure time delays for single components of the real-time system [Behr et al., 2015] but can also be used to analyse if and when a particular station contributed to an earthquake detection. Doing this for Switzerland, we found that only in very few cases were P-wave detections missed by the closest stations for magnitudes >= ??. Further, our analysis of expected alert times for every participating network confirms that data transfer times are the dominant source of delay in most networks [NOTE: run delayeew for every network]. The latter was done using the method to predict expected alert times proposed by Behr et al. [2015], which applies a Monte Carlo simulation to the distributions of observed systems delays. Using the same approach but assuming ideal, that is current state of the art, data communication delays and an EEW algorithm capable of predicting ground motion levels based on two P-wave detections only, we can estimate the maximum possible earthquake alerts. Figure 4c shows that for such a configuration alerts could be sent to areas expecting very strong shaking even for moderate sized earthquakes with magnitudes > 5.8. While estimating magnitude and location of an earthquake from the very first P-wave recordings at only two sensors, several algorithms have already demonstrated its feasibility [Reference PRESTo, Onsite, Gutenberg Algorithm, Virtual Seismologist].
